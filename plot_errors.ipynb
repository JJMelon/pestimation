```python
import matplotlib.pyplot as plt
import pickle

# Load the error results from the pickle file
with open('error_results.pkl', 'rb') as f:
    errors = pickle.load(f)

# Extract data for plotting
noise_std = [entry['std'] for entry in errors]  # Noise standard deviations
state_estimation_errors = [entry['nmae_voltA_real'] for entry in errors]  # State estimation errors
parameter_estimation_errors = [entry['nae_line_length'] for entry in errors]  # Parameter estimation errors
```

```python
# Plot State Estimation Error (Box and Whisker)
plt.figure(figsize=(12, 6))
plt.boxplot(state_estimation_errors, positions=noise_std, widths=0.075, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='blue'), 
            medianprops=dict(color='red'))
plt.title("State Estimation Error vs Noise")
plt.xlabel("Noise Standard Deviation (%)")
plt.ylabel("State Estimation Error (%)")
plt.grid(axis='y')
plt.xticks(noise_std)  # Ensure x-ticks match noise levels
plt.tight_layout()
plt.show()
```

```python
# Plot Parameter Estimation Error (Box and Whisker)
plt.figure(figsize=(12, 6))
plt.boxplot(parameter_estimation_errors, positions=noise_std, widths=0.075, patch_artist=True, 
            boxprops=dict(facecolor='lightgreen', color='green'), 
            medianprops=dict(color='red'))
plt.title("Parameter Estimation Error vs Noise")
plt.xlabel("Noise Standard Deviation (%)")
plt.ylabel("Parameter Estimation Error (%)")
plt.grid(axis='y')
plt.xticks(noise_std)  # Ensure x-ticks match noise levels
plt.tight_layout()
plt.show()
```
